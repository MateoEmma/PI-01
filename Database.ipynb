```python
dfD = pd.read_csv('Datasets/drivers.json')
dfD.to_csv('drivers.csv')
dfD.head()
```

```python
del dfD['Unnamed: 0']
```

```python
dfD.head()
```

```python
dfD['name'] = dfD['name'].replace({'{' : ''}, regex=True)
dfD['name'] = dfD['name'].replace({'}':''}, regex=True)
```

```python
dfD[['forename','surname']] = dfD.name.str.split(pat= ',', expand= True)
```

```python
del dfD['name']
```

```python
dfD['forename'] = dfD['forename'].replace({'forename':''}, regex=True)
dfD['surname'] = dfD['surname'].replace({'surname':''}, regex=True)
dfD.head()
```

```python
dfD['forename'] = dfD['forename'].replace({'"',''})
dfD['surname'] = dfD['surname'].replace({'"',''})
dfD.head()
```

```python
dfD[['a','forename']] = dfD.forename.str.split(pat= ':', expand= True)
dfD.head()
```

```python
dfD[['a','surname']] = dfD.surname.str.split(pat= ':', expand= True)
del dfD['a']
dfD.head()
```

```python
dfD.to_csv('drivers.csv')
```
